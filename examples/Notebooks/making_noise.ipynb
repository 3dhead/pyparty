# Creating SEM test data

```python
from pyparty import Canvas, Grid
from pyparty.bundled import load_style
load_style('louic')
```

Our task is to generate some test data for SEM images that can be used to test various segmentation algorithms.  We'd like to test how well the algorithms perform in segmenting particles at in the presence of uneven contrast, noise and various degrees of particle overlap.  First, let's define some constants to help reuse this scripts

```python
RESOLUTION = (512, 512)   #Image resolution
BGGRAY = 190                     #Light gray background
PCOLOR = 255                      #White particles
NOISE_DENSITY = 0.30        #percent
```

#### Step 1: Create a light-gray background; add particles

```python
c = Canvas()
c.rez = RESOLUTION
c.background = BGGRAY

for (cx, cy) in c.gpairs('centers'):
    c.add('circle', radius=10, center=(cx,cy), color=PCOLOR)
    
c.show();
```

#### Step 3: Add a constant and uneven contrast gradient using "Grid".

**Note**: We use any2rgb to RGB-convert the background so that plotting doesn't try to autonormalize the data.

```python
def warped(xx, yy, power=1.0):
    """ 2d function to generate uneven contrasts """
    return abs(np.cos(xx**power) + np.sin(yy**power))

g1 = Grid(style='h')
gradientbg = g1.blend_bg(cgray.image, weight=0.35)
gradientbg = any2rgb(gradientbg)

g2 = Grid()
g2.set_zfcn(warped)
warpedbg = g2.blend_bg(cgray.image, weight=0.25)
warpedbg = any2rgb(warpedbg)

ax0, ax1, ax2 = splot(1,3, figsize=(15,15))

showim(c.image, ax0)
showim(gradientbg,  'gray', ax1, title='Do I want to keep blend?');
showim(warpedbg, 'gray', ax2);
```

#### Step 4: Add some noise

```python
ax0, ax1, ax2 = splot(1,3, figsize=(15,15))

# Make noise module for pyparty (this, saltpaper)
def whitenoise(img, coverage=NOISE_DENSITY):
    """ Requires 3-channel image; adds white noise"""

    noise = np.random.random(img.shape[0:2])
    img[noise > 1- coverage] = (1,1,1)
    return img

showim(whitenoise(c.image) , 'gray', ax0)
showim(whitenoise(gradientbg) , 'gray', ax1)
showim(whitenoise(warpedbg) , 'gray', ax2);
```
