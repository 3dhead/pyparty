### Invert Gray Image

```python
def compare(img):
    ax1, ax2, ax3 = splot(1,3)

    showim(img, ax1)
    showim(invert(img), ax2)
    showim(img + invert(img), ax3, title='sum');

    print img.min(), img.max()
    print invert(img).min(), invert(img).max()
```

```python
from pyparty.utils import invert
from pyparty import splot, showim

compare(c.grayimage)
```
