### Introduction

*pyparty* installs a script called **objecthunt**.  The purpose of objecthunt is to separate labeled objects from an image, and output common statistics, plots and perform common pyparty operations.  This is best shown by example, so consider the pyparty test image, *test_labeled*:

```python
from pyparty.data import test_labeled
from skimage.io import imsave

REZ = (1000, 1024)

IMAGE = test_labeled(REZ)
imshow(IMAGE);
#imsave('PATH/test_image.png', IMAGE);
```

**objecthunt** will automate a series of common  *pyparty* [Canvas](http://nbviewer.ipython.org/github/hugadams/pyparty/blob/master/examples/Notebooks/basictests.ipynb?create=1) and [Multicanvas](http://nbviewer.ipython.org/github/hugadams/pyparty/blob/master/examples/Notebooks/multi_tutorial.ipynb?create=1) operations including:
  
   - Pie, Histograms and multiplots of various species
   - Particle counts, statistics
       - Includes attributes stats such as min, max and mean circularity of each group.
   - Image conversions, Gray, Binary etc.. deconstructed by species.
   
All the various outputs will be discussed later in the *Configuration* section

### Getting started

First, test that *objecthunt* is installed properly by typing:
   - ``objecthunt - h``
   
This should output a [argparser](https://docs.python.org/dev/library/argparse.html) help message.

The only *positional* arguments are an **image** and an **out directory**.  The default configuration is found in the config file in the pyparty/scripts/objecthunt folder [view it here](https://github.com/hugadams/pyparty/blob/master/pyparty/scripts/objectmodeling/_configobjhunt.py).

At this point, I have saved the cropped test_image above into my home directory.  Let's check out the output with the default configuration:

   - ``objecthunt ~/test_image.png ~/foo_out -p ignore=black``
   
Where we set a parameter, *ignore* to black as the test image has a black background.  Alternatively, this could have been set directly in the config file. <font color='blue'>**Any run parameter can be overwritten at the command line.**</font>

### Configuration

There are several configurable parameters, mostly having to do with plotting styles, and which analysis to do and which not to.  Hopefully, the variable names are self-explantory.  Here are just a few examples:

- **multidir** Directory name of where to output multicanvas plots (pie, hist etc..)
- **multihist** = 'hist.png' : **multihistkwds** = {}
    - Name of multicanvas piechart and any valid keywords (e.g. bins=50)
- **ignore** = 'white'
    - Background color to ignore in images

**<font color='blue'>For plots and directories, setting value to None (e.g. multihist=None) will remove that part of the analysis</font>**.  For example, setting ``canvasdir = None`` will avoid the canvas-by-canvas analysis entiely!

Another key parameter is the **mapper**, which allows the user to specify group names and colors of objects.  For our example images, the mapper would be set as follows:

    mapper = [
        ('singles','red'),
        ('doubles',(0,1,0)),
        ('trimers', 'gold' ),
        ('clusters', 'magenta')
        ]
        
Where the colors are specified as any valid pyparty color style.  *The script will guess the closest color from the one specified, so exact colors are not required!*  The order of the names in the mapper persists in the plots!  Here is an example of the *pie* and *multishow* when the above mapper is used in the config file:

```python
from pyparty.tools.arraytools import unique
from pyparty.utils import any2rgb

unique( any2rgb(IMAGE) )
```
