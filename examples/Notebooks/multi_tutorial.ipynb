#### Initialization (from list of canvases (*canvii*) )

Let's use "Canvas.random_circles()* to generate 10 canvases with a random number of circles between 1 and 50 and random names.  The random name generator was taken from [this thread](http://stackoverflow.com/questions/2257441/python-random-string-generation-with-upper-case-letters-and-digits).

MultiCanvas must be initialized with a **names** and **canvii** instance; both must be equally-sized lists.  We will show alternative constructors subsequently.

```python
import string
import random 
from pyparty import MultiCanvas

def rand_name(size=6, chars=string.ascii_lowercase + string.digits):
     return ''.join(random.choice(chars) for _ in range(size))

names, canvii = [], []
for i in range(10):
    names.append( rand_name() )
    canvii.append( Canvas.random_circles(n=random.randint(1,100)) )
    
mc = MultiCanvas(names=names, canvii=canvii)
mc
```

#### Sorting

Let's sort this (by name).  When sorting multicanvas, <font color='red'>**do not use pythons sorted() function; used mc.sort()**</font>

```python
mc.sort(inplace=True)
mc
```

#### Indexing

Indexing is as you'd expect.  Valid indicies include:
   - Integer
   - Integer slice
       - eg [5:3], [3::] etc...
   - Single name (string)
   
A slice that returns multiple objects will return a MulitCanvas.  A slice that returns a single element will return a Canvas.  This is the same behavior as Python's list slicing.

```python
print mc[4:7]

NAME = mc.names[0]
mc[NAME].show(title='Showing Canvas ``%s"' % NAME);
```

Setting a deleting by item are also supported, **but only by single key**; the same as a python dictionary:

```python
mc['test'] = Canvas.random_circles(n=20)
print mc['test']
del mc['test']
```
