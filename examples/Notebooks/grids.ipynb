## pyparty Grids

`pyparty` provides some basic Grid classes, the goals of which are to provide a gridding/tiling system for creating particles and background.  As with other aspects of pyparty, grids can be created from barebones:

http://stackoverflow.com/questions/19586828/drawing-grid-pattern-in-matplotlib

`pyparty` Grids wrap numpy.meshgrid() and provide a basic API for human-natural interaction.  For example, grids retain information on their spacing, tiles, centers, corners etc...  Before looking at the Grid classes in general, let's look at the builtin **CartesianGrid** of the canvas class.  First, let's conjur up some random circles.  Canvas has a special constructor for this:

```python
from pyparty import Canvas
c=Canvas.random_circles(n=15, bgcolor='honeydew')
```

Canvas' grid can be accessed as an attribute:  

```python
c.show(grid=True)
c.grid
```

We can change the tiling via the **spacing** or **points** parameters, for both the x and y dimensions.  *Spacing* sets the width of each grid tile; *points* let the users fix the grid to a preset number of spaces.  At any time, the **reset_grid()** method will restore the default resolution and spacing.  

```python
from pyparty import splot

ax1, ax2 = splot(1,2)

c.grid.xpoints=4
c.show(ax1, gcolor='black', title='4 "X" divisions')
c.reset_grid()

c.grid.yspacing=10
c.show(ax2, gcolor=(1,0,0), title='10 pixels per "Y" division')
c.reset_grid();
```

The "grid" keyword to *show()* takes a variety of arguments:
    
   - True or "grid" : Full grid
   - "centers" : tile center dots
   - "corners" : tile bottom right corner dots
   - "xlines" : horizonal grid lines
   - "ylines" : vertical

```python
axs = splot(2,2)

c.show(axs[0], grid='centers', title='centers')
c.show(axs[1], grid='corners', title='bottom-left corners')
c.show(axs[2], grid='hlines', title='hlines')
c.show(axs[3], grid='vlines', title='vlines')

# Zoom in to for clarity
for ax in axs:
   ax.set_xlim(0,100)
   ax.set_ylim(0,100)
```

There's not much magic going on here, becuase *centers*, *corners* etc.. are all accessible indicies in the underlying grid.  We could access them directly and use them as masks on any old array.

```python
#Make a blank 512x512 array (ones=white)
blank = np.ones((512,512,3))

#Get the indicies of corners, centers
blank[c.grid.centers] = (1,0,0)
blank[c.grid.corners] = (0,0,1)
imshow(blank)
plt.xlim(0,100)
plt.ylim(0,100);
```

What if you don't want the bottom right corner?  The **grid.grid_points()** method is nice for returing the center and various 4 corner postions via ints:
  
  - 0 or 'c'  : centers
  - 1 or 'br' : bottom right corners
  - 2 or 'bl' : bottom left corners
  - 3 or 'tl' : top left corners
  - 4 or 'tr' : top right corners 

```python
from pyparty import to_normrgb as rgb
from pyparty import showim 
#zeros for black bg (Make 600x600 so translated corners can extend past the image)
blank = np.zeros((600,600,3))
gp = c.grid.grid_points

#Get the indicies of corners, centers
blank[gp('c')] = rgb('red')
blank[gp('tl')] = rgb('blue')
blank[gp('bl')] = rgb('green')
blank[gp('tr')] = rgb('yellow')
blank[gp('br')] = rgb('orange')

showim(blank, title='First tile (otherwise; orange overwrites)')
plt.xlim(0,36)
plt.ylim(36,0);
```

### Mapping Particles

A major motivator for the canvas grid is to map particles to the grid points.  For example, if we want to map a circle to every third circle:

```python
from skimage.data import lena

cnew=Canvas(background=lena(), rez=(512, 400))

for (cy, cx) in zip(*cnew.grid.grid_points('center')):
    cnew.add('circle', radius=5, center=(cx,cy), color='black')
    
for (cy, cx) in zip(*cnew.grid.grid_points('bottomright')):
    cnew.add('ellipse', xradius=5, yradius=10, center=(cx,cy), color='white')

cnew.show(gcolor='teal');
```

## Grids for backgrounds

Grids were also developed to create backgrounds for both testdata, and just for psychedelic backgrounds.  While canvas implements an implicit instance of a CartesianGrid; it is mostly for mapping particles.  The other Grids are more flexible for creating interesting backgrounds for testdata.

The **Grid** class, unlike the TiledGrid or the CartesianGrid, will plot as a continuous 2d function.  Use the **.zz** attribute to access the ndarray.

```python
from pyparty.tools.grids import Grid, TiledGrid

ax1, ax2 = splot(1,2)

g = Grid()
tg = TiledGrid(xpoints=20, ypoints=20)
ax1.imshow(g.zz)
ax2.imshow(tg.zz)
tg
```

### numpy meshgrids

Grid() is insensitive to divisions or tiles; it is a continuous 2d function.  Every grid first starts by making a mesh that returns two arrays, XX and YY.  **The numpy API explains this in full**, but here's an example of how the above mesh is generated from XX, YY:

```python
x=np.linspace(0,1,512)
y=np.linspace(0,1,512)
XX, YY = np.meshgrid(x,y)

ax1, ax2 = splot(1,2)

ax1.imshow(XX, plt.cm.autumn)
ax2.imshow(YY, plt.cm.autumn);
```

Notice that by plotting only XX or only YY, we get a different result.  In effect, we can define an arbitrary function that takes in (XX, YY) as parameters, and returns ZZ.  So, for the first plot, it would look something like:

```python
def justXX(xx, yy):
    return xx

def add_em(xx, yy):
    return xx+yy

ax1, ax2 = splot(1,2)

ax1.imshow(justXX(XX,YY))
ax2.imshow(add_em(XX,YY));
```

### pyParty Grid --> zfcn

Grid() implements a **zfcn** attribute to incorporate mapping a 2d function to the grids.  *zfcn* is an arbitrary function **that must accept**  XX and YY positional argument.  

Because it's very common to want to be able to generate a backdrop of changing gradiant, Grid() accepts a *fade* parameter that will call an implicit zfcn that performs combinations of horizontal, vertical and diagonal gradients (ie, all combinations of the above examples), and has a *reverse* option

```python
ax1, ax2 = splot(1,2)

tg1 = TiledGrid(zfcn='fade', style='d')
tg2 = TiledGrid(zfcn='fade', style='d', reverse=True)

ax1.imshow(tg1.zz, plt.cm.bone)
ax2.imshow(tg2.zz);
```

### Custom Z function

Let's define a custom z function that takes the cos of xx and sin of yy and set to our regular, and non tiled grids.  I'll also play with some of the grid colors for cool effect.

```python
ax1, ax2 = splot(1,2)

def ztrigfcn(xx, yy):
    return abs(np.cos(xx**2) + np.sin(yy**2))

tg.zfcn = ztrigfcn
g.zfcn = ztrigfcn

ax1.imshow(g.zz, plt.cm.Blues)
ax2.imshow(tg.zz, plt.cm.Blues);
```

Notice that the function is applied on the grid **BEFORE** tiling.  If you convert to integer during your zfunction, you will implictly create a tiled grid.  Probably best to avoid this!

```python
def zsquared_int(xx, yy):
    return (xx*3).astype(int) + yy.astype(int)

g_foo = Grid(zfcn=zsquared_int)
showim(g_foo.zz, plt.cm.gray, title="Induced tiling through zfunction");
```

### Grids as canvas backgrounds.  

A grid can directly be passed into a canvas as the background:

```python
ax1, ax2 = splot(1,2)

c.background=g
c.show(ax1, plt.cm.Blues)
c.reset_background()

c.background=tg1
c.show(ax2, plt.cm.jet);
```
