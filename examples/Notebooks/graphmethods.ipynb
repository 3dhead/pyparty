#### Step 2: Import images, convert to gray, plot

#### Step ?: Non-graph based segmentation

Use adaptive threshold routine from scikit image wich computes regions in local neighborhoods (blocksize) around each pixel.

```python
from skimage.filter import threshold_adaptive

axrows = splot(3,3,figsize=(15,15), flatten=False)

BLOCKSIZES = [15, 30, 45]

for idx, img in enumerate(IMAGES):
    row = axrows[idx] # Row 1, 2 or 3
    for j in range(3):
        block = BLOCKSIZES[j]
        bin_img = threshold_adaptive(img, block, offset=0)
        showim(bin_img, row[j], 'gray', 
               title='%s; block = %s' % (NAMES[idx], block));
```

While we didn't fully explore the parameter space of **adaptive**, we see that it is fairly insensitive to contrast gradients.  This is fairly expect; local methods are *local* in this regard (see below).  But we do see that the noise really does a bad job.

#### Graph Cuts:

 Adapted from http://scikit-learn.org/stable/auto_examples/cluster/plot_segmentation_toy.html

```python
from sklearn.feature_extraction import image
from sklearn.cluster import spectral_clustering
from pyparty.utils import crop

CROPPED = crop(SALTY, (0,0,200,200)) 
mask = crop(bin_clean, (0,0,200,200))

graph = image.img_to_graph(CROPPED, mask=mask)

# Take a decreasing function of the gradient: we take it weakly
# dependent from the gradient the segmentation is close to a voronoi
graph.data = np.exp(-graph.data / graph.data.std())

# Force the solver to be arpack, since amg is numerically
# unstable on this example
labels = spectral_clustering(graph, eigen_solver='arpack')
```

```python
label_im = -np.ones(mask.shape)
label_im[mask] = labels

#imshow(CROPPED)
imshow(label_im);
```
