**n-circ** deals with special orientations of connected circles.  These are especially useful to my research in particular, but we will continue to work towards generalized multi-particle support.

#### Let's draw some circles, lines, and ellipses

```python
c.add('circle')

c.add('ellipse', center=(300,300), yradius=50, xradius=30, color='red' )
c.add('ellipse', center=(100,400), yradius=20, xradius=100, color=128, phi=10.0 ) 
c.add('ellipse', center=(400,250), yradius=200, xradius=40, color='g', phi=10.0 ) 

c.add('line', length=200, center=(200,100))
c.add('line', length=400, width=5, color='purple', phi=45.0)

c.show(title='Lines and ellipses');
print c.particles
```

#### Bezier curve and polygons

```python
c.clear_particles()
c.add('square')
c.add('square', center=(200,300), color='teal', length=50)
c.add('square', xverts=(300,300,400,400), yverts=(300,400,400,300),
      phi=40.0)

c.add('triangle', length=50, color='red', center=(450, 100), 
      phi=90.0)
c.add('triangle', xverts=(300, 350, 400), yverts=(200, 100, 200), 
      color='y')

c.add('bezier', ystart=20, ymid=400, yend=500, xstart=50, 
      xmid=90, xend=200)
c.add('bezier', ystart=200, ymid=300, yend=400, xstart=50,
      xmid=150, xend=500, weight=0.2, color=(0,1,0))

c.add('polygon', xverts=(100,100, 200, 200), yverts=(100,200, 100, 200), 
      color='magenta', phi=45.0)
c.add('polygon', sets=( (200,200), (200, 250), (300, 200), (300,250)),
      color = 'gray', orientation=90.0)

c.show(title='Bezier and polygons');
```

You probably noticed that each shape has several constructors, especially polygons.  Under the hood, **Polygon** is actually examining its args (center, length etc...) and trying to pick the correct constructor.  The four constructors for Polygon are:

- **Polygon.from_length**
   - Polygon.from_length( center=(x,y), length=50 )
- **Polygon.from_length_width**
   - Polygon.from_length_width( center=(x,y), length=50, width=30 )
- **Polygon.from_sets**
   - Polygon.from_sets( sets=(x1, y1, x2, y2) )
- **Polygon.from_verts**
   - eg Polygon.from_verts( xverts=(x1,x2), yverts=(y1,y2) )

Each polygon subclass implements these differently.  For example, **Square** will error if it recieves a *width* argument; this is built into *Square.from_length_width*.  These constructors try to be intelligent about storing default values; for example, passing a "center" argument to Square will result in "from_length()" construction at a default length specified in the config file.  At the end of the day, ``pyparty`` tries to be flexible in its constructors, and hopefully enough examples are provided to ascertain all of the available call types.

#### Multi-circles

``pyparty`` shapes infrastructure was designed with composite shapes in mind; however, at this time, only provies an API for multi circles.  

```python
c.available('n-circle')
```

Dimers are a custom-particle for drawing neighboring circles.  This is a multiparticle built around a call to skeleton pattern elements in pyparty.pattern.elements, which returns skeleton coordinates (verticies) at which *Dimer()* draws circles of specified radii.

```python
c.clear_particles()
c.add('tetramer', 2,8,14,4, center=(250, 250), color=(1,0,0))
c.add('trimer', r1=30, center=(100,100), color=(.2,.2,.2))
c.add('tetramer', radius_1=20, center=(400, 400), color=(0, .5, 0))
c.show();
```

Notice that the particle manager only has 3 particles, indicating that each bundle is a *single* particle:

```python
len(c)
```

Don't forget that we can use **pmap()** to apply a function to all of the particles.  For example, if I want to change the orientation of the particles by 45 degrees clockwise, and swap the center's x and y coordinates, I merely define a function that changes the attribute of each particle and returns the particle:

```python
XMAX = c.rx

def reorient(p):
    """ Rotate shape 45 degrees; reflect cx """
    p.cx = XMAX-p.cx
    p.phi = p.phi + 45.0
    return p

c.pmap(reorient, inplace=True)
c.show();
```

In addition to **orientation (ie, phi)**, one may also customize these particles via **overlap** keyword, as well as adding other radii (IE radius_2, radius_3, radius_3).  

```python
c.clear_particles()

c.add('dimer', radius_1=40, radius_2=30, center=(400, 100), 
      phi=45, color=(0,1,0))
c.add('trimer',   radius_1=40, center=(100,100), overlap=0.2, 
      color=(.7,.2,.2) )
c.add('tetramer', radius_1=40, center=(100, 400), overlap=0.2, 
      color=(0, .5, 0) )
c.add('tetramer', radius_1=40, radius_2=20, radius_4=35, 
      center=(400, 400), color=(0, .5, .3) )

c.show();
```

**Note:** Notice the bottom right square looks rather funky.  Pyparty generates the square by first generating an X-pattern of verticies (pyparty.pattern.elements.simple) and then for each segment of the X, it shortens or lengths it based on the user-set radius.  Not all combinations of circle sizes can result in perfectly spaced entities, which is why square 4 looks stranghts, or broken in the above plot.

```python
c.clear_particles()
c.background='black'

c.add('ellipse', center=(250,250), yradius=210, xradius=50, 
      color='green', phi=20.0 ) 
c.add('ellipse', center=(250,250), yradius=200, xradius=40, 
      color='yellow', phi=20.0 ) 

c.show(title='Accidental Corn Plot');
```

The texture of the corn is actually due to the way indicies are rounded when rotating; this will be fixed soon.  These artifacts don't appear in matplotlib patches.  There also seems to be a bug in matplotlib.patches; ellipses are undersized...
