```python
from pyparty.tools.canvas import Canvas

c=Canvas()
c.available()
```

### Let's draw some lines and some ellipses

```python
c.add('line', ystart=5, yend=500, xstart=5, xend=500, color=(1,0,0))
c.add('line', ystart=15, yend=300, xstart=50, xend=500, color=(0,1,0))
c.add('line', ystart=200, yend=500, xstart=300, xend=500, color=(0,0,1))
c.add('ellipse', center=(300,300), yradius=50, xradius=20, color=(1,0,0) )
c.add('ellipse', center=(100,400), yradius=20, xradius=100, color=(0,1,0) )

ax=c.show()
ax.axes.set_title('Lines and ellipses');
```

### Bezier curve and polygons

```python
c.clear_particles()
c.add('bezier', ystart=20, ymid=400, yend=500, xstart=50, xmid=90, xend=200)
c.add('bezier', ystart=200, ymid=300, yend=400, xstart=50,
      xmid=150, xend=500, weight=0.2, color=(0,1,0))
c.add('polygon', xcoords=(100,200, 200, 100), ycoords=(200,300, 300, 200), color=(1,0,1))

ax=c.show()
ax.axes.set_title('Bezier and polygons (polygons not working)');
```

### Dimers, Trimers and Squares

Dimers are a custom-particle for drawing neighboring circles.  This is a multiparticle built around a call to skeleton pattern elements in pyparty.pattern.elements, which returns skeleton coordinates (verticies) at which *Dimer()* draws circles of specified radii.

```python
c.clear_particles()
c.add('square', radius_1=50, center=(250, 250), color=(1,0,0))
c.add('trimer', radius_1=30, center=(100,100), color=(.2,.2,.2) )
c.add('square', radius_1=20, center=(400, 400), color=(0, .5, 0) )

c.show();
```

Notice that the particle manager only has 3 particles:

```python
c.count
```

This allows us to operate on the multiparicle itself.  For example, if I want to change the orientation of the particles by 45 degrees clockwise, I merely define a function that changes the attribute of each particle and returns the particle:

```python
def reorient(p):
    p.orientation = p.orientation + 45.0
    return p

c.pmap(reorient, inplace=True)
c.show();
```

```python
c.clear_particles()

c.add('dimer', radius_1=40, radius_2=30, center=(400, 100), 
      orientation=60.0, color=(0,1,0))

c.add('trimer', radius_1=40,  overlap=0.2, center=(100,100), color=(.2,.2,.2) )
c.add('square', radius_1=40, center=(100, 400), overlap=0.2, color=(0, .5, 0) )
c.add('square', radius_1=40, radius_2=20, radius_4=35, center=(400, 400), color=(0, .5, .3) )

c.show();
```
