### Multi-circles

``pyparty`` shapes infrastructure was designed with composite shapes in mind; however, at this time, only provies an API for multi circles.  

```python
c.available('n-circle')
```

Dimers are a custom-particle for drawing neighboring circles.  This is a multiparticle built around a call to skeleton pattern elements in pyparty.pattern.elements, which returns skeleton coordinates (verticies) at which *Dimer()* draws circles of specified radii.

```python
c.clear_particles()
c.add('tetramer', radius_1=50, center=(250, 250), color=(1,0,0))
c.add('trimer', radius_1=30, center=(100,100), color=(.2,.2,.2) )
c.add('tetramer', radius_1=20, center=(400, 400), color=(0, .5, 0) )
c.show();
```

Notice that the particle manager only has 3 particles, indicating that each bundle is a *single* particle:

```python
len(c)
```

Don't forget that we can use **pmap** to apply a function to all of the particles.  For example, if I want to change the orientation of the particles by 45 degrees clockwise, and swap the center's x and y coordinates, I merely define a function that changes the attribute of each particle and returns the particle:

```python
XMAX = c.rx

def reorient(p):
    """ Rotate shape 45 degrees; reflect cx """
    p.cx = XMAX-p.cx
    p.phi = p.phi + 45.0
    return p

c.pmap(reorient, inplace=True)
c.show();
```

In addition to **orientation (ie, phi)**, one may also customize these particles via **overlap** keyword, as well as adding other radii (IE radius_2, radius_3, radius_3).  

```python
c.clear_particles()

c.add('dimer', radius_1=40, radius_2=30, center=(400, 100), 
      phi=45, color=(0,1,0))

c.add('trimer', radius_1=40,  overlap=0.2, center=(100,100), color=(.7,.2,.2) )
c.add('tetramer', radius_1=40, center=(100, 400), overlap=0.2, color=(0, .5, 0) )
c.add('tetramer', radius_1=40, radius_2=20, radius_4=35, center=(400, 400), color=(0, .5, .3) )

c.show();
```

**Note:** Notice the bottom right square looks rather funky.  Pyparty generates the square by first generating an X-pattern of verticies (pyparty.pattern.elements.simple) and then for each segment of the X, it shortens or lengths it based on the user-set radius.  Not all combinations of circle sizes can result in perfectly spaced entities, which is why square 4 looks stranghts, or broken in the above plot.
