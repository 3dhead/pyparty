### Let's draw some circles, lines, and ellipses

```python
c.add('circle', center=(400,100), radius=40, color=(.5, .3, .2))
c.add('ellipse', center=(300,300), yradius=50, xradius=30, color='red' )
c.add('ellipse', center=(100,400), yradius=20, xradius=100, color=40 ) 

c.add('line', ystart=5, yend=500, xstart=5, xend=500, color='aqua')
c.add('line', ystart=15, yend=300, xstart=50, xend=500, color=(0,1,0))
c.add('line', ystart=200, yend=500, xstart=300, xend=500, color=(0,0,1))

c.show(title='Lines and ellipses');
```

### Bezier curve and polygons

```python
c.clear_particles()
c.add('bezier', ystart=20, ymid=400, yend=500, xstart=50, xmid=90, xend=200)
c.add('bezier', ystart=200, ymid=300, yend=400, xstart=50,
      xmid=150, xend=500, weight=0.2, color=(0,1,0))
c.add('polygon', xcoords=(100,200, 200, 100), ycoords=(200,300, 300, 200), color=(1,0,1))

ax=c.show()
ax.axes.set_title('Bezier and polygons (polygons not working)');
```

### Dimers, Trimers and Squares

Dimers are a custom-particle for drawing neighboring circles.  This is a multiparticle built around a call to skeleton pattern elements in pyparty.pattern.elements, which returns skeleton coordinates (verticies) at which *Dimer()* draws circles of specified radii.

```python
c.clear_particles()
c.add('tetramer', radius_1=50, center=(250, 250), color=(1,0,0))
c.add('trimer', radius_1=30, center=(100,100), color=(.2,.2,.2) )
c.add('tetramer', radius_1=20, center=(400, 400), color=(0, .5, 0) )

c.show();
```

Notice that the particle manager only has 3 particles, indicating that each bundle is a *single* particle:

```python
len(c)
```

This allows us to operate on the multiparicle itself.  For example, if I want to change the orientation of the particles by 45 degrees clockwise, I merely define a function that changes the attribute of each particle and returns the particle:

```python
def reorient(p):
    """ Rotate shape 45 degrees """
    p.orientation = p.orientation + 45.0
    return p

c.pmap(reorient, inplace=True)
c.show();
```

In addition to **orientation**, one may also customize these particles via **overlap** keyword, as well as adding other radii (IE radius_2, radius_3, radius_3).  

```python
c.clear_particles()

c.add('dimer', radius_1=40, radius_2=30, center=(400, 100), 
      orientation=60.0, color=(0,1,0))

c.add('trimer', radius_1=40,  overlap=0.2, center=(100,100), color=(.7,.2,.2) )
c.add('tetramer', radius_1=40, center=(100, 400), overlap=0.2, color=(0, .5, 0) )
c.add('tetramer', radius_1=40, radius_2=20, radius_4=35, center=(400, 400), color=(0, .5, .3) )

c.show();
```

**Note:** Notice the bottom right square looks rather funky.  Pyparty generates the square by first generating an X-pattern of verticies (pyparty.pattern.elements.simple) and then for each segment of the X, it shortens or lengths it based on the user-set radius.  Not all combinations of circle sizes can result in perfectly spaced entities, which is why square 4 looks stranghts, or broken in the above plot.
