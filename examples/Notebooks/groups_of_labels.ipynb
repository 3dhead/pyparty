## Multiple Labels from a Segmented Image

This example is inspired by a real task in the lab.  We obtained a carefully segmented image of gold nanoparticles and wanted to quantify various aspects of each group (ie single particles, dimers, trimers and large clusters).  For example, how does the eccentricity vary with each subgroup?  

Before we can answer these questions, let's look at the image:

```python
from pyparty import Canvas, splot, showim
from pyparty.data import nanolabels, nanogold

NANOLABELS = nanolabels()
ax1, ax2 = splot(1,2, figsize=(12,6))

showim(nanogold(), ax1, 'gray', title='SEM nanoparticles')
showim(NANOLABELS, ax2, 'spectral', 
       title='size-segmented nanoparticles');

print 'unique colors:', np.unique(NANOLABELS)
```

### Working with Masks

The segmented image has five unique labels, with 0 being the background.  The first task is to split the segmented image into masks for each of the particle categories.  ``pyparty`` provides a utility to simplify this task, **multi_mask(labeledimage, names, ignore=0, sort=True)**:

- labeledimage : image of labels color uniquely by particle group (see above)
- names (optional) : Names for each category.  If none, actual value of image is used.
    - For example, "singles" isntead of (1.0), which is the pixel value in the image
- ignore : Pixel values to ignore
    - Default to 0, which is a black background.  
- sort : Return masks in a OrderedDict vs. unorderd standard dictionary
    - OrderedDict requires python.collection (standard with 2.7 or higher)

multi_mask will return a dictionary of boolean arrays, one corresponding to each subgroup of labels.  This is best illustrated with an example.

```python
from pyparty.multi import multi_mask, multi_labels

axes = splot(2,2, figsize=(10,6))

masks = multi_mask(NANOLABELS, 'singles', 'dimers', 'trimers', 'clusters')
for idx, (name, image) in enumerate(masks.items()):
    showim(image, axes[idx], 'gray', title=name)

type(masks), masks.keys()[0], masks.values()[0].shape
```

We see that the return of *multi_masks()* is a dictionary of boolean arrays, each one corresponding to subgroup of interest.  For certain tasks, the masks are sufficient; for example, the area of the image occupied by dimers can be found by the ratio of white pixels to dark pixels:

```python
L, W = NANOLABELS.shape
dimer_area = sum(masks['dimers'])

print 'Dimer coverage: %.1f percent' % \
    (100.0 * ( dimer_area / (L * W)) )
```

### Working with Particle Labels

For more complex tasks, its better to read the masks into pyparty.  The **multi_labels()** function is designed to take a dict or list of masked arrays and return a dict / list of either **Particles** or **Canvas** objects.  The arguments of multi-labels are:

- **as_canvas** : bool (True)
   - return a Canvas for each masked array vs. just Particles
- **prefix** : bool or str
   - Prefix ascribed to particle names (eg dimer_0, dimer_1); if True, names from multi_mask dictionary keys are used (see example below)
- **neighbors** : 4 or 8 (4)
   - Connectivity of labels
- **pmankwargs** : args passed directly to ParticleManager.from_labels() (eg pmin / pmax)

Let's first deal with a dictionary of *Particles* (ie **as_canvas = False**)

```python
nps = multi_labels(masks, as_canvas=False)
nps.keys()[0], nps.values()[0][0:5]
```

We see that multi_labels is still an ordered dictionary, but instead of masks, we have **ParticleManager** objects.  Had we used **as_canvas**, we'd have a full Canvas for each value of the dictionary.  Notice that **prefix=True** auto-filled the particle NAMES to that of the original key.  Let's pursue the following task:

*Plot the distribution of areas in each particle subgroup(ie histogram)?  And how does this compare to multiples of the **mean single particle** area?*

The actual computation of the necessary quantities is quite easy.  Let's go ahead and separate the area of each subgroup and get the mean of the *single particles area* 

```python
AREAS = [p.area for p in nps.values()] 
GROUPNAMES = nps.keys()
mu = np.mean(AREAS[0])
```

The histogram takes a bit more effort; mostly for bells and whistles.  *Note that I'm going to leave the large clusters out of the histogram because some of the clusters have such a large area that it screws with the aesthetics of the histogram bins*

```python
# Plot constants
YMAX = XMAX = 300
BINS=30
COLOR = ['r','g','y','magenta']

# Histogram (EXCLUDE CLUSTERS - areas[3])
plt.hist((AREAS[0:3]), bins=30, label=GROUPNAMES, 
         color=COLOR[0:3], stacked=True)

# Add vlines to separate mean area
plt.vlines((mu, 2*mu, 3*mu), 0, YMAX)

# Add text to plot
amu = '$A_\mu$'
textkwds = {'size':13, 'color':'blue',
            'bbox':{'facecolor':'gray', 'alpha':.5}}

plt.text(25, 275, 'x < %s' % amu, **textkwds)
plt.text(80, 275, '%s < x < 2%s' % (amu, amu), **textkwds)
plt.text(158, 275, '2%s < x <2%s' % (amu,amu), **textkwds)
plt.text(250, 200, 'x > 3%s' % amu, **textkwds)

# Set title, limits, axis labels
plt.xlim(0,XMAX)
plt.title('Particle Area Distribution (%s bins)' % BINS, size=19)
plt.ylabel('Counts', size=17)
plt.xlabel('Area', size=17)

plt.legend();
```

We see that the single particles distribution is centered around the single particle mean area nicely.  The histogram is misleading because it displays the **number** of particles moreso than their relative portion of the surface coverage.  For this, a pie chart is more apt.  First, let's get the net area of each particle group, as well as the net count

```python
netcount = [len(area) for area in AREAS]
netarea = [sum(area) for area in AREAS] 
```

Again, the rest of the tedium comes down to tweaking a plot; the pie chart API is pretty easy.

```python
ax1, ax2 = splot(1,2, figsize=(12,6))

chartkwds = {'shadow':True,'autopct':'%1.1f%%','colors':COLOR}

ax1.pie(netcount, labels=GROUPNAMES, **chartkwds)
ax2.pie(netarea, labels=GROUPNAMES, explode=(0,0,0,0.1), **chartkwds)

ax1.set_title("Number distribution", size=15)
ax2.set_title("Area distribution", size=15);
```

We see that although clusters only account for 2.6% of the particle count, they make up over 10% of the coverage.

### Working with Canvas Labels

Let's go back to "multi_labels()" and use **as_canvas=True**.  This is particularly useful if you need full-fledged Canvas operations.  At anytime, the particles are directly accessible through **c.particles**, so the above results would still be easy to obtain.

```python
nps = multi_labels(masks, as_canvas=True)
nps.keys()[0], nps.values()[0]
```

We can really make the dimers pop by using canvas patchshow() for example

```python
c_dimers = nps['dimers']
c_dimers.background = nanogold()
c_dimers.patchshow(pmap='jet');
```

The canvas also makes it quite easy to access the original particle mask through the **pbinary** attribute, or to apply thresholding directly on the background image:

```python
ax1, ax2 = splot(1,2, figsize=(12,6))

c_dimers.show(ax1, 'pbinary', title='Original dimer masks')

# Remove the background and then compute adaptive threshold on particles
c_dimers.background='black'
c_dimers.set_threshfcn('adaptive', block_size=3)
showim(c_dimers.binaryimage, ax2, 'gray',
       title='Dimer masks under poor thresholding');
```

This contrived example shows how a poor thresholding function would degrade particle masks; really it just shows how mapping masks to canvas can extend functionality.
