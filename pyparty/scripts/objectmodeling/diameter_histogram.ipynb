#### Step 3: Obtain arrays for the area, equivalent diameter and particle circularity.  

While equivalent diameter is actually a builtin descriptor, since we are going to change units from pixels to nanometers (through SCALE), it's trivial to define the equivalent diameter as:

$d = \sqrt{\frac{4A}{\pi}}$

The circularity we will borrow from ImageJ's definition:
http://rsbweb.nih.gov/ij/docs/guide/146-30.html

$4\pi \times \frac{\mbox{Area}}{\left(\mbox{Perimeter}\right)^2}$


```python
areas = clab.area 
areas_nm = areas * SCALE**2

d_nm = np.sqrt(4 * areas_nm  / np.pi) #PIXELS
r_nm = d_nm / 2.0
circ = np.pi *( ( 4 * areas) / (clab.perimeter)**2 )
```

##### Step 5: Git centers of bins, use them to fit a guassian

This uses several utilities found in *scripts.fitutils*

- **get_bin_points():** return left, right or center values of bins
- **hist_max():** Find the maximum histogram bin.  Can specify start and stop for plots that are likely to have local maxima.  Returns index, x and y point.
- **psuedo_symmetric():**
- **optimize_guassian():**

```python
from pyparty.scripts.fitutils import get_bin_points, hist_max

bincenters = get_bin_points(bin_edges, position='c')
idxmax, xmax, ymax = hist_max(counts, bincenters, idx_start=1, idx_stop=None)
'Max bin: index=%i, diameter=%.1f, count=%1f' % (idxmax, xmax, ymax)
```

Explanation

```python
from pyparty.tools.arraytools import nearest
from pyparty.scripts.fitutils import psuedo_symmetric, optimize_gaussian

idx_left = nearest(bincenters, (xmax * (1.0-LLEFT) ) )[0]
symm_counts, symm_centers=psuedo_symmetric(counts, bincenters, idx_start=idx_left)          
        
### Fit a shorten, optimized gaussian to the data ###
histfit, fit_amp, fit_mean, fit_sig=optimize_gaussian(symm_counts, symm_centers)

def best_fit(x):
    """ Based on fit parameters, builds a guassian """
    return fit_amp*np.exp(-(x-fit_mean)**2/(2.*fit_sig**2))
      
plt.plot(bincenters, best_fit(bincenters), color='black', ls='--');
plt.scatter(bincenters, best_fit(bincenters));
best_fit(bincenters)
showhist();
```

```python
about_zero=0.2 #Lower lim on fit function where it's regarded as 0
dx =fit_sig / 100.0 #Number of increments between mean and sigma used when iterating on histogram

def fit_min_max():
    f=best_fit #For convience rereference function
       
    ### Set start, end and increment for iterating over fit fcn
    dmax=d_nm.max()
    dmin=d_nm.min()
    
    ### Set starting values for x and y
    xmin=xmax=fit_mean
    y=f(fit_mean)
        
    ### Iterate from mean to x=dmax
    while y > about_zero:
        xmax=xmax+dx
        y=f(xmax)
            
        if xmax > dmax:
            xmax=None
            break

    y=f(fit_mean)
            
    ### Iterate from mean to x=dmin
    while y > about_zero:
        xmin=xmin-dx
        y=f(xmin)
            
        if xmin < dmin:
            xmin=None
            break
        
    return (xmin, xmax)

patches=showhist()    
fmin, fmax = fit_min_max()
xfunc = np.arange(fmin, fmax, dx)
plt.plot(xfunc, best_fit(xfunc), color='blue');

plt.title('Guassian Fit Range   %.1f - %.1f (%s)'%
          ( xmax*(1.0 - LLEFT), xmax, UNIT))

for patch in patches[idx_left:idxmax+1]:
    patch.set_color('pink') #Lots of ways to customize this, like doing stripes/border etc.

### Add text about guassian moments                     
xpos=.6 * float(plt.xlim()[1])
ypos=.7 * float(plt.ylim()[1])
plt_txt='$ \mu=$%.1f nm \n $\sigma=$%.1f nm\n x=%.1f nm\n y=%s counts' \
               % (fit_mean, fit_sig, xmax,  ymax )
plt.text(xpos, ypos, plt_txt, fontsize=12) 

plt.vlines(fmin, 0.0, ymax, color='r', linestyles='dashed')  
plt.vlines(fmax, 0.0, ymax, color='r', linestyles='dashed')    
plt.vlines(2.0*fit_mean, 0.0, ymax, color='g', linestyles='dashed'); 
```

#### Step ?: BSA Counting

Bsa is counted based on an experimental fit to a dataset from NIST.  (CITE)
Based on the diameter of the nanoparticle, it estimates the number of BSA particles on the nanoparticle.  The function bsa_count returns the bsa fill fraction, that is, the bsa per unit surface area for that particle.

```python
from pyparty.scripts.BSA import bsa_count

fill_frac = bsa_count(d_nm, style='dual')
```

The fill fraction at 100% coverage is $0.0386\mbox{nm}^2$.  If we multiple the fill_frac by the particle surface area, we will get the number of BSA per molecule.  For singles, we compute the area simply from the equivalent diameter, assuming a circle.  For aggregates and other species, we need to model the surface area.

```python
def rint(x): return int(round(x,0))

surf_area = 4.0 * np.pi * r_nm**2
bsa = fill_frac * surf_area
'Total Nps: %s / Total bsa: %s' % (len(clab), rint(bsa.sum()))
```
