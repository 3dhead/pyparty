```python
from pyparty import Canvas, splot, showim
from pyparty.bundled import load_style
load_style('plain')
```

Let's define some constants, such as the size ranges for NOISE and LARGE_AGG (aggregates).  Other ranges like the upper limit for single particle are determined dynamically based on the mean particle size and fit.

```python
SCALE = 2.78    #nm / pixel
BINS = 37       # Histogram bins (hueristic)
NOISEMAX = 10   #nm (upper limit for noise)
LARGE_AGG = 150 #nm diameter equiv for large aggregates
```

#### Step 1: Load a pre-thresholded image and set it as canvas background

```python
from pyparty.data import nanobinary
c=Canvas(background=nanobinary())
c.show();
```

#### Step 2: Assign labels from the background

```python
clab = c.from_labels(exclude ='white', pbinary=False, pmin=8, neighbors=4) 
clab.show(title='Particles from labels')
clab.particles
```

The steps for creating a canvas of particle labels are a bit wonky, because there are two different scenarios.  The first is that we have a canvas that already has particles (eg we added 50 circles).  In this case, the binarized image is obtained from c.pbinary.  The other case, which we have here, is that the canvas doesn't have particles, and we want the background image to be the source of the particles.  That is what the **pbinary=False** option does.  This is explained in detail in the separate tutorials on labels.

#### Step 3: Obtain arrays for the area, equivalent diameter and particle circularity.  

While equivalent diameter is actually a builtin descriptor, since we are going to change units from pixels to nanometers (through SCALE), it's trivial to define the equivalent diameter as:

$d = \sqrt{\frac{4A}{\pi}}$

The circularity we will borrow from ImageJ's definition:
http://rsbweb.nih.gov/ij/docs/guide/146-30.html

$4\pi \times \frac{\mbox{Area}}{\left(\mbox{Perimeter}\right)^2}$


```python
areas = clab.area 
areas_nm = areas * SCALE**2

d_nm = np.sqrt(4 * areas_nm  / np.pi) #PIXELS
r_nm = d_nm / 2.0
circ = np.pi *( ( 4 * areas) / (clab.perimeter)**2 )
```

#### Step 4: Plot a histogram; cut data between diameters of 10-100nm

```python
dcut = d_nm[(d_nm>10) & (d_nm<100)]

plt.hist(dcut, 37, color='orange');
plt.xlim(10,100)
plt.xlabel('Equivalent Diameter')
plt.ylabel('Counts');
```

#### Step ?: BSA Counting

Bsa is counted based on an experimental fit to a dataset from NIST.  (CITE)
Based on the diameter of the nanoparticle, it estimates the number of BSA particles on the nanoparticle.  The function bsa_count returns the bsa fill fraction, that is, the bsa per unit surface area for that particle.

LATER, DO THIS WITH SINGLES, DIMERS ETC>> SEPARATELY

```python
from pyparty.scripts.BSA import bsa_count

fill_frac = bsa_count(d_nm)
```

The fill fraction at 100% coverage is $0.0386\mbox{nm}^2$.  If we multiple the fill_frac by the particle surface area, we will get the number of BSA per molecule.  For singles, we compute the area simply from the equivalent diameter, assuming a circle.  For aggregates and other species, we need to model the surface area.

```python
def rint(x): return int(round(x,0))

surf_area = 4.0 * np.pi * r_nm**2
bsa = fill_frac * surf_area
'Total Nps: %s / Total bsa: %s' % (len(clab), rint(bsa.sum()))
```
