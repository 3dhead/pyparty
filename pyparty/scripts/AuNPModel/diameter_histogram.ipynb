```python
pylab.rcParams['figure.figsize'] = 8, 5.5
import matplotlib.pyplot as plt
import numpy as np
import pylab as P

from skimage import data, img_as_float
from skimage import io
from skimage.morphology import label
from pyparty import Canvas, splot, showim
```

```python
%cd ~/Desktop/f2_b1_100k21_low
filename = 'f2_b1_100k21_low_adjusted.tif'
img = io.imread(filename)
c = Canvas()
c.background = img
```

```python
ax1, ax2 = splot(1,2)

showim(c.image, ax1, 'gray', title='Low threshold image');

labels = label(img)
showim(labels, ax2, 'jet', title='Labels');
```

```python
print labels.shape
print np.unique(labels)
labels
```

```python
clab = c.from_labels(exclude ='white', pbinary=False, pmin=10, neighbors=4) 
clab.show(title='Particles from labels')
clab.particles
```

```python
a = clab.area * 2.78**2     #PIXELS
d = np.sqrt(4 * a / pi) #PIXELS
dcut = d[(d>10) & (d<100)]
```

```python
from math import pi
plt.hist(dcut, 37);
plt.xlim(10,100)
```
